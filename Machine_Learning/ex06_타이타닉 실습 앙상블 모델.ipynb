### 전처리된 파일 불러오기

```python
import pandas as pd
X_train = pd.read_csv('X_train.csv')
y_train = pd.read_csv('Y_train.csv')
X_test = pd.read_csv('X_test.csv')
```

```python
X_train.shape, y_train.shape, X_test.shape
```

### 앙상블 모델 사용하기
- 머신러닝에서 성능이 좋은 모델 
- 여러개의 DecisionTree 모델을 사용 - RandomForest 모델 

### RandomForest 모델 불러오기 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forest_model = RandomForestClassifier(n_estimators=50)
# 이번에 사용할 dicision tree의 50개를 통해서  randomforest모델을 만들겠다. 
```

```python
# warning을 무시하고 출력하기
import warnings
warnings.filterwarnings('ignore')
```

```python
# 교차검증 ( cv= 5) 
from sklearn.model_selection import cross_val_score
score = cross_val_score(forest_model, X_train, y_train, cv=5).mean()

# 학습 
forest_model.fit(X_train,y_train)
# 평가
forest_model.score(X_train,y_train) # 과대적합 되어있음,
# 예측 결과를 csv 파일로 내보내기 
pre = forest_model.predict(X_test)  # 예측할때는 검증용(test) 파일로 측정
gender_sub = pd.read_csv('./data/titanic/gender_submission.csv') # 점수 예측 후 기존의 성별 생존율과 비교
gender_sub['Survived'] = pre
# kaggle에서 업로드해서 점수 확인해보기
gender_sub.to_csv('mysub03.csv',index=False)
# 점수 상승됨. 
```

### 하이퍼 파라미터 튜닝

```python
forest_model1 = RandomForestClassifier(n_estimators=100)
```

```python
# 하이퍼 파라미터 튜닝  
# n_estimators(트리의 개수) =[50, 70, 90, 110, 130,150],
# max_depth(깊이 최대치) = [3,4,5,6,7],
# max_features(특성) = [0.3, 0.4, 0.5, 0.6, 0.7]

```

```python
# 하이퍼 파라미터를 딕셔너리형태로 저장(param)
param = {'n_estimators' : [50, 70, 90, 110, 130,150],
         'max_depth' :[3,4,5,6,7],
          'max_features' : [0.3, 0.4, 0.5, 0.6, 0.7]}
```

```python
# GridSearchCV 적용 
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(RandomForestClassifier(), param, cv=5, verbose=2) 
# verbose : 학습하는 과정을 보여줌 
```

```python
# 찾아낸 최적의 조합을 적용한 모델로 학습, 평가, 예측 --> csv파일로 내보내기
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```
